```python
import pandas as pd

# Load dataset
data = pd.read_csv("marketing_data.csv")

# 1️⃣ Unknown values ko "NaN" me convert karte hain taake unko handle kar saken
data.replace("unknown", pd.NA, inplace=True)

# 2️⃣ Missing values count (check karne ke liye)
print("Missing values after converting 'unknown':")
print(data.isnull().sum())

# 3️⃣ Simple filling strategy — missing values ko 'Unknown' ya mode se fill karte hain
categorical_cols = ['job', 'marital', 'education', 'contact', 'poutcome']
for col in categorical_cols:
    data[col].fillna(data[col].mode()[0], inplace=True)

# 4️⃣ Encoding binary columns (yes/no) into 0/1
binary_cols = ['default', 'housing', 'loan', 'deposit']
for col in binary_cols:
    data[col] = data[col].map({'yes': 1, 'no': 0})

# 5️⃣ One-Hot Encoding for non-binary categorical columns
data = pd.get_dummies(data, columns=['job', 'marital', 'education', 'contact', 'month', 'poutcome'], drop_first=True)

# ✅ Final check
print("\nFinal dataset shape after encoding:", data.shape)
print("Sample preview:")
print(data.head(3))
```

```python
# ===============================
# Full Setup & Visual Insights - Safe Version
# ===============================

# 1️⃣ Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

# 2️⃣ Load Dataset
data = pd.read_csv("marketing_data.csv")

# 3️⃣ Clean & Encode
data.replace("unknown", pd.NA, inplace=True)

# Fill missing categorical values with mode
categorical_cols = ['job','marital','education','contact','poutcome']
for col in categorical_cols:
    data[col] = data[col].fillna(data[col].mode()[0])

# Encode binary columns
binary_cols = ['default','housing','loan','deposit']
for col in binary_cols:
    data[col] = data[col].map({'yes':1,'no':0})

# One-Hot Encoding for categorical variables except 'job' (used for safe plot)
data_encoded = pd.get_dummies(data, columns=['marital','education','contact','month','poutcome'], drop_first=True)

# 4️⃣ Visual Insights

# ----- 1. Deposit vs Age -----
plt.figure(figsize=(8,5))
sns.histplot(data=data, x='age', hue='deposit', multiple='stack', bins=20)
plt.title('Deposit by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# ----- 2. Deposit vs Job -----
job_deposit_counts = pd.crosstab(index=data['job'], columns=data['deposit'])
job_deposit_counts.plot(kind='barh', stacked=True, figsize=(10,5), color=['salmon','skyblue'])
plt.title('Deposit Count by Job Type')
plt.xlabel('Number of Customers')
plt.ylabel('Job')
plt.show()

# ----- 3. Balance Distribution -----
plt.figure(figsize=(8,5))
sns.boxplot(x='deposit', y='balance', data=data)
plt.title('Balance Distribution by Deposit')
plt.show()

# ----- 4. Previous Outcome Effect -----
plt.figure(figsize=(8,5))
sns.countplot(x='poutcome', hue='deposit', data=data)
plt.title('Deposit by Previous Campaign Outcome')
plt.show()

# ----- 5. Call Duration vs Deposit -----
plt.figure(figsize=(8,5))
sns.boxplot(x='deposit', y='duration', data=data)
plt.title('Call Duration vs Deposit')
plt.show()
```
